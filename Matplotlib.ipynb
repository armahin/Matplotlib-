# **3D Surface Plot**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
zz = xx**2 + yy**2
fig = plt.figure(figsize = (15,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,zz,cmap='plasma')
fig.colorbar(p)
```

# **Contour Plot**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
zz = xx**2 + yy**2
fig = plt.figure(figsize = (15,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,zz,cmap='viridis',alpha = 0.7)
fig.colorbar(p)
```
